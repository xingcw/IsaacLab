```python
import numpy as np
import matplotlib.pyplot as plt

# Configuration
proximity_threshold = 0.2
eps_tanh = 0.001

# Compute k
k = 2 * proximity_threshold / np.log(2.0 / eps_tanh - 1)  # â‰ˆ 0.0263

# Distance range
d = np.linspace(0, 2*proximity_threshold, 500)

# Convergence reward
conv = 1 - np.tanh(d / 0.1)

# Plot
plt.figure()
plt.plot(d, conv)
plt.xlabel('Distance to Goal')
plt.ylabel('1 - tanh(distance / k)')
plt.title('Convergence Reward vs. Distance')
plt.grid(True)
plt.show()
```
